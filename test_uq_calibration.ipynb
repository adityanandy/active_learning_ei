```python
def new_obj(theta, dists, epsilons):
    L = 0
    for i in range(len(epsilons)):
        epsilon = epsilons[i]
        dist = dists[i]
        L += -np.log(norm.pdf(epsilon, 0, np.sqrt(dist*theta**2)))
    return L
```

```python
cost = [new_obj(x, dists, epsilons) for x in np.linspace(1, 3, 100)]
```

```python
trace0 = go.Scatter(
    x=np.linspace(1, 3, 100),
    y=cost,
    mode='markers',
    opacity=0.5,
    marker=dict(
        size=4,
        color='black',
    ),
    line=dict(color='black', width=2, dash='dash'),
)
data = [trace0]
layout = go.Layout()
layout.update(glob_layout)
layout["xaxis"].update({'title': 'dists'})
layout["yaxis"].update({'title': 'epsilons'})
layout.legend.update(x=0, y=1.0, bgcolor='rgba(0,0,0,0)')
layout.update(xaxis=dict(), yaxis=dict())
layout.update(height=500, width=500, showlegend=False)

fig = dict(data=data, layout=layout)
iplot(fig)
```
